### Taking Apart the SEC Feed XML document

```python
root = ET.parse('..//data//source//sec//ia-firm//IA_FIRM_SEC_Feed_02_28_2022.xml').getroot()
```

```python
# Initialize df column lists
dict_items = {}
for firm in root[0]:
    for item in firm:  # Info, MainAddr, etc.
        if item.tag == 'FormInfo':
            for subitem in item:  # Part1A
                for item_n in subitem:  # Item 1, Item 2A, etc.
                    if item_n.tag not in dict_items.keys():
                        dict_items[item_n.tag] = {}
                    for q in item_n.attrib.keys():
                        if q not in dict_items[item_n.tag].keys():
                            dict_items[item_n.tag][q] = []

# ACTUALLY populate df column lists
firmcrdnb = []
busnms = []
legalnm = []
umbrrgstn = []

strt1 = []
strt2 = []
city = []
state = []
cntry = []
postlcd = []
phnb = []

for firm in root[0]:  # [:3] REMEMBER TO REMOVE INDEX UP TO 10 WHEN READY FOR FULL FILE
    for item in firm:  # Info, MainAddr, etc.
        if item.tag == 'Info':
            firmcrdnb.append(item.attrib['FirmCrdNb'])
            busnms.append(item.attrib['BusNm'])
            legalnm.append(item.attrib['LegalNm'])
            try:
                umbrrgstn_text=item.attrib['UmbrRgstn']
                umbrrgstn.append(umbrrgstn_text)
            except:
                umbrrgstn_text=''
                umbrrgstn.append(np.nan)
        elif item.tag == 'MainAddr':
#             strt1.append(item.attrib['Strt1'])
            try:
                strt1_text=item.attrib['Strt1']
                strt1.append(strt1_text)
            except:
                strt1_text=''
                strt1.append(np.nan)
#             strt2.append(item.attrib['Strt2'])
            try:
                strt2_text=item.attrib['Strt2']
                strt2.append(strt2_text)
            except:
                strt2_text=''
                strt2.append(np.nan)
#             city.append(item.attrib['City'])
            try:
                city_text=item.attrib['City']
                city.append(city_text)
            except:
                city_text=''
                city.append(np.nan)
#             state.append(item.attrib['State'])
            try:
                state_text=item.attrib['State']
                state.append(state_text)
            except:
                state_text=''
                state.append(np.nan)
#             cntry.append(item.attrib['Cntry'])
            try:
                cntry_text=item.attrib['Cntry']
                cntry.append(cntry_text)
            except:
                cntry_text=''
                cntry.append(np.nan)
#             postlcd.append(item.attrib['PostlCd'])
            try:
                postlcd_text=item.attrib['PostlCd']
                postlcd.append(postlcd_text)
            except:
                postlcd_text=''
                postlcd.append(np.nan)
#             phnb.append(item.attrib['PhNb'])
            try:
                phnb_text=item.attrib['PhNb']
                phnb.append(phnb_text)
            except:
                phnb_text=''
                phnb.append(np.nan)
        elif item.tag == 'FormInfo':
            for subitem in item:  # Part1A
                for item_n in subitem:  # Item 1, Item 2A, etc.
                    for q, qval in dict_items[item_n.tag].items():
                        if q in item_n.attrib.keys():
                            dict_items[item_n.tag][q].append(item_n.attrib[q])
                        else:
                            dict_items[item_n.tag][q].append(np.nan)
```

```python
# Unpack the horrible dictionary of dictionaries into a nice list of lists (ha)

ls_col_names = []
ls_cols_lists = []
for item, qdict in dict_items.items():
    for q, ls_q in qdict.items():
        ls_col_names.append(q)
        ls_cols_lists.append(ls_q)
        
ls_final = [firmcrdnb] + [busnms] + [legalnm] + [umbrrgstn] + [strt1] + [strt2] + [city] + [state] + [cntry] + [postlcd] + [phnb] + ls_cols_lists
```

```python
# Convert list of lists into an inverted dataframe
df_init = pd.DataFrame(ls_final)

# Make the dataframe nice
df_final = df_init.T
df_final.columns = ['firmcrdnb'] + ['business_name'] + ['legalnm'] + ['umbrrgstn'] + ['strt1'] + ['strt2'] + ['city'] + ['state'] + ['cntry'] + ['postlcd'] + ['phnb'] + ls_col_names
```
